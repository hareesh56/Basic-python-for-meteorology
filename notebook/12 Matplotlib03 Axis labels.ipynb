```python
y = np.random.randn(24) # temperature during every hour in a day
x = np.arange(0,len(y)) # time in a day 0UTC --> 23UTC

fig, axes = plt.subplots(1, 1, figsize=(10, 5))

print(len(x)-len(y))

plt.plot(x, y, 'r',label="Temperature")
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')


axes.set_xlabel("Time (h)", labelpad=5, fontsize=18, fontname='serif', color="k")
axes.set_ylabel("Temperature $^oC$", labelpad=15, fontsize=18, fontname='serif', color="m")
axes.set_title("Matplotlib is cool", loc='right', fontsize=16, fontname='serif', color="b")
plt.legend()
fig.tight_layout()
#fig.savefig("matplotlib_labels.pdf", dpi=300)
```
