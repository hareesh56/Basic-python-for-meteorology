```python
# create some artificial data
time = np.linspace(0, 24, 60)
observed  = np.random.randn(60)
simulated = np.random.randn(60)
```

```python
print(observed.shape)
print(observed)
print('======')
print(simulated.shape)
```

```python
# plot figure
fig, ax1 = plt.subplots(figsize=(8, 4))

# plot 1st figure
ax1.plot(time, observed, lw=2, color='blue')

# set attribute for 1st figure
ax1.set_title('observed and simulated temp.', fontsize=22)
ax1.set_xlabel('times (min.)', fontsize=20)
ax1.set_ylabel('observed temp ($^oC$)', fontsize=14, color='blue')

# set color for the 1st figure 
for label in ax1.get_yticklabels():
    label.set_color('blue')

# TO MAKE TWIN AXIS, we need the below command
ax2 = ax1.twinx()

# plot 2nd y axis
ax2.plot(time, simulated, lw=2, color='red')
ax2.set_ylabel('simulated temperature ($^oC$)', fontsize=14, color='red')

# set color for the 1st figure 
for label in ax2.get_yticklabels():
    label.set_color('red')
     
fig.tight_layout()
```
