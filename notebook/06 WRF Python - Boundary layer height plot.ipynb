# 1. Import libraries

```python
# Loading necessary libraries
import numpy as np
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature

from wrf import (to_np, getvar, smooth2d, get_cartopy, cartopy_xlim,
                 cartopy_ylim, latlon_coords, interplevel)
```

# 2. Download data

```python
# specify where is the location of the data
path_in = "data/"
path_out = "./"
# Open the NetCDF file
ncfile = Dataset(path_in + 'wrfout_d01_2016-05-09_00^%00^%00')
```

# 3. Take out the variables

```python
# Get the boundary layer height
PBLH = getvar(ncfile, "PBLH")
print(PBLH.dims)
```

# 4. Plotting

```python
PBLH.plot()
```
