```python
# import python packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['font.family'] = 'serif'
```

```python
# create artificial function y
x = np.linspace(-4, 14, 1000)
y = 1/(1 - x**2) + 10/(2 + ((7 - 2*x)/0.1)**2)
```

```python
# plot figure
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y, linewidth=2, c='r')

ax.set_xlabel('X', fontsize=15)
ax.set_ylabel('Y', fontsize=15)
```

```python
# plot figure
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y, linewidth=2,c='r')
ax.set_xlabel('X', fontsize=15)
ax.set_ylabel('Y', fontsize=15)

x0, x1 = 3, 4
ax.axvline(x0, ymin=0.25, ymax=0.75, color="k", linestyle=":")
ax.axvline(x1, ymin=0.25, ymax=0.75, color="k", linestyle=":")
```

```python
# plot figure
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y, linewidth=2,c='r')
ax.set_xlabel('X', fontsize=15)
ax.set_ylabel('Y', fontsize=15)

# just to indicate the region of interest
x0, x1 = 3, 4
ax.axvline(x0, ymin=0.25, ymax=0.75, color="k", linestyle=":")
ax.axvline(x1, ymin=0.25, ymax=0.75, color="k", linestyle=":")

# x and y values in the region of interest
x_inner = np.linspace(x0, x1, 1000)
y_inner = 1/(1 - x_inner**2) + 10/(2 + ((7 - 2*x_inner)/0.1)**2)

#[left, bottom, width, height]
ax2 = fig.add_axes([0.53, 0.49, 0.35, 0.35], facecolor='white')
ax2.plot(x_inner, y_inner, linewidth=2,c='b')
ax2.set_xlabel('X subset', fontsize=10, c='b')
ax2.set_ylabel('Y subset', fontsize=10, c='b')
```
