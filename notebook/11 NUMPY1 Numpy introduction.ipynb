## 5. Mesh grid array by Numpy

The numpy.meshgrid function is used to create a grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing

```python
x = np.arange(-180, 181, 60)
```

```python
y = np.arange(-90, 91, 30)
```

```python
print(x)
print(y)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
print(X)
print(X.shape)
plt.imshow(X)
```

```python
print(Y)
print(Y.shape)
plt.imshow(Y)
```

```python
temperature = (X + Y)
```

```python
print(temperature)

cmap = plt.get_cmap('PiYG')
fig, (ax0) = plt.subplots(nrows=1)
im = ax0.pcolormesh(X, Y,temperature, cmap=cmap)
fig.colorbar(im, ax=ax0)
```
