```python
import glob
```

# listing all of the file in a folder named "data"

```python
for file_name in glob.glob('data/*'):
    print(file_name)
```

# listing all of the "NetCDF" file in a folder named "data"

```python
for file_name in glob.glob('data/*.nc'):
    print(file_name)
```

# store all netcdf file path into a list 

```python
list_netcdf = []
for file_name in glob.glob('data/*.nc'):
    list_netcdf.append(file_name)
```

```python
# print first and last netcdf file
print('the first netcdf file:', list_netcdf[0])
print('the last netcdf file:', list_netcdf[-1])
```
