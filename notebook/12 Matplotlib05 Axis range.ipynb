```python
# create some artificial data
y = np.random.randn(24) # temperature during every hour in a day
x = np.arange(0,len(y)) # time in a day 0UTC --> 23UTC
```

```python
# plot figure
fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(15, 4))

# 1st figure
ax1.plot(x, y, c='b', lw=2)
ax1.set_xlim(0, 24)
ax1.set_ylim(-4, 4)
ax1.set_title('our defined values')

# 2nd figure
ax2.plot(x, y, c='b', lw=2)
ax2.axis('tight')
ax2.set_title('tight-axis figure')

# 3rd figure
ax3.plot(x, y, c='b', lw=2)
ax3.axis('equal')
ax3.set_title('set equal xlim and ylim')

# create vertical and horizontal lines
ax1.axvline(x=0, color='k')
ax1.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.axhline(y=0, color='k')
ax3.axvline(x=0, color='k')
ax3.axhline(y=0, color='k')
fig.tight_layout()
```
