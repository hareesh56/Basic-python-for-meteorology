### Boolean arrays and conditional expressions

```python
a = np.arange(0,11,1)
print(a)
```

```python
b = np.arange(11,0,-1)
print(b)
```

```python
a < b
```

```python
np.all(a < b)
```

```python
np.any(a < b)
```

```python
for i in range(len(a)):
    if a[i] < b[i]:
        print('a[i] = ', a[i],' is smaller than b[i] = ', b[i])
    else: 
        print('a[i] is either greater or equal b[i]')
```

```python
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
```

```python
x > 0
```

```python
1 * (x > 0) # return 0 and 1
```

```python
x * (x > 0) # return values of x at those its values are true
```

```python
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
print(x)
```

```python
np.where(x < 0, x**2, x**3)
```

```python
np.select([x < -1, x < 2, x >= 2],
          [x**2  , x**3 , x**4])
```

```python
np.choose([2, 0, 0, 1, 1, 1, 2, 2, 2], 
          [x**2,    x**3,    x**4])
# 3 different powers, then select 1 option [index 0, or 1 or 2] to apply on the values
```

```python
abs(x)
```

```python
x[abs(x) > 2]
```

```python
np.nonzero(abs(x) > 2) # return index not values
```

```python
x[np.nonzero(abs(x) > 2)] # return values
```
