## 2. Index and slicing

### 2.1) One dimensional array in Numpy

```python
temperature_ = np.arange(24, 48, 1)
```

```python
temperature_
```

```python
print(temperature_.shape)
```

```python
temperature_[0]  # the first element
```

```python
np.round(temperature_[23], 2) # the last element
```

```python
temperature_[4]  # the fifth element, at index 4
```

```python
temperature_[0:2] # the first two element at index 0 and 1
```

```python
temperature_[0:24] # all elements 
temperature_[:]
```

```python
temperature_[0:-1] # the first to (the last - 1)
```

```python
temperature_[1:-1:2] # the second to (the last - 1), with incremental step of 2
```

```python
temperature_[0:24:6] # the first to (the last), with incremental step of 6
```

```python
temperature_[:5] # the first five elements
```

```python
temperature_[-5:] # the last five elements
```

```python
temperature_[::-2] # from the last to the first, in reverse order, with incremental step of -2
```

```python
temperature_[::-6] # from the last to the first, in reverse order, with incremental step of -6
```

## 2.2) Multidimensional dimensional array in Numpy

```python
FA = lambda m, n: n + 10 * m #N Column, M rows
#Column = column + 10 * row
print(FA)
```

```python
A = np.fromfunction(FA, (6, 6), dtype=int)
```

```python
A
```

```python
A[:, 2]  # the third column
```

```python
A[2, :]  # the third row
```

```python
A[:3, :3]  # upper left half diagonal block matrix
```

```python
A[:3, -3:]  # upper right left off-diagonal block matrix
```

```python
A[-3:, :3]  # lower left off-diagonal block matrix
```

```python
A[-3:, -3:]  # lower right half diagonal block matrix
```

```python
A[::2, ::2]  # every second element starting from 0, 0
```

```python
A[1::2, 1::2]  # every second element starting from 1, 1
```

## 2.3) More about arrays

- Be careful when assign a current array to a new variable. 
- The original array can be modified too

# 2.3.1) first example

```python
B = A[1:5, 1:5]
print(B)
```

```python
B[:, :] = 0
print(B)
```

```python
print(A)
```

# 2.3.2) second example

```python
C = B[1:3, 1:3]
print(C)
```

```python
C[:, :] = 99
print(C)
```

```python
print(B)
```

# 2.3.3) to solve, use copy()

```python
B[:,:] = 0
print(B)
C = B[1:3, 1:3].copy()
```

```python
print(C)
```

```python
C[:, :] = 1  
# We now assign value 1 for C
# this does not affect B since we use ".copy()". and C is a copy of the frame B[1:3, 1:3]
```

```python
print(C)
```

```python
print(B) # Should be all zeros
```

## 2.4) Boolean index

```python
A = np.linspace(0, 1, 11)
print(A)
```

```python
A[2]
```

```python
A[[0, 2, 4, 6]]
```

```python
A > 0.5 # if A > 0.5 will return True, otherwise, False (logical array)
```

```python
A[A > 0.5] # only return values of A those are larger than 0.5
```

## 3. Reshape and resize

```python
data1 = np.array([[1, 2], [3, 4], [5, 6]])
print(data1)
print('the shape of data1 is:', data1.shape)
```

```python
data2 = np.reshape(data1, (1, 6))
print(data2)
print('the shape of data2 is:', data2.shape)
print('the #dim of data2 is:', data2.ndim)
```

```python
data3 = data1.reshape(6)
print(data3)
print('the shape of data3 is:', data3.shape)
print('the #dim of data3 is:', data3.ndim)
```
