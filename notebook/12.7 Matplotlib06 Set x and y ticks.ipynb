```python
# x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
print(x.shape)
# y values
y = np.sin(x) * np.sin(2*x) * np.exp(-x**1/5)

# plot figure
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(x, y, color='m',lw=3) 
ax.set_title("Default ticks", fontsize=20)

fig.tight_layout()
print('x min:', np.round(np.min(x),1))
print('x max:', np.round(np.max(x),1))
print('y min:', np.round(np.min(y),1))
print('y max:', np.round(np.max(y),1))
```

```python
# x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
print(x.shape)
# y values
y = np.sin(x) * np.sin(2*x) * np.exp(-x**1/5)

# plot figure
fig, ax = plt.subplots(1, 2, figsize=(10, 3))

ax[0].plot(x, y, color='m',lw=3)
ax[0].set_title("default ticks")

ax[1].plot(x, y, color='m',lw=3)
ax[1].set_yticks([-3, 0, 3])
ax[1].set_xticks([-6,-3, 0, 3, 6])
ax[1].set_title("set_xticks")

fig.tight_layout()
```

```python
# x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
print(x.shape)
# y values
y = np.sin(x) * np.sin(2*x) * np.exp(-x**1/5)

# plot figure
fig, ax = plt.subplots(1, 3, figsize=(10, 3))

ax[0].plot(x, y, color='m',lw=3)
ax[0].set_title("default ticks")

ax[1].plot(x, y, color='m',lw=3)
ax[1].set_yticks([-3, 0, 3])
ax[1].set_xticks([-6, 0, 6])
ax[1].set_title("set_xticks")

ax[2].plot(x, y, color='m',lw=3)
ax[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-4, 0, 4]))
ax[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
# ax[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax[2].xaxis.set_major_locator(mpl.ticker.FixedLocator([-8, 0, 8]))
ax[2].xaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
ax[2].set_title("set_major_locator")
ax[2].set_yticks(np.arange(-4,4.5,0.5))
ax[2].set_xticks(np.arange(-8,9,1))

ax[0].grid()
ax[1].grid()
ax[2].grid()
fig.tight_layout()
```

```python
# x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
print(x.shape)
# y values
y = np.sin(x) * np.sin(2*x) * np.exp(-x**1/5)

# plot figure
fig, ax = plt.subplots(1, 4, figsize=(12, 3))

ax[0].plot(x, y, color='m',lw=3)
ax[0].set_title("default ticks")

ax[1].plot(x, y, color='m',lw=3)
ax[1].set_yticks([-3, 0, 3])
ax[1].set_xticks([-6, 0, 6])
ax[1].set_title("set_xticks")

ax[2].plot(x, y, color='m',lw=3)
ax[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-4, 0, 4]))
ax[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(16))
# ax[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax[2].xaxis.set_major_locator(mpl.ticker.FixedLocator([-8, 0, 8]))
ax[2].xaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
ax[2].set_title("set_major_locator")
ax[2].set_yticks(np.arange(-4,4.5,0.5))
ax[2].set_xticks(np.arange(-8,9,1))

ax[3].plot(x, y, color='m',lw=3)
ax[3].set_yticks([-4, 0, 4])
ax[3].set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
ax[3].set_xticklabels(['$-2\pi$', '$-\pi$', 0, r'$\pi$', r'$2\pi$'])
ax[3].xaxis.set_minor_locator(mpl.ticker.FixedLocator([-3 * np.pi / 2, -np.pi/2, 0, np.pi/2, 3 * np.pi/2]))
ax[3].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
ax[3].set_yticks(np.arange(-4,5,1))
ax[3].set_title("set_xticklabels")

ax[0].grid()
ax[1].grid()
ax[2].grid()
ax[3].grid()

fig.tight_layout()
```
