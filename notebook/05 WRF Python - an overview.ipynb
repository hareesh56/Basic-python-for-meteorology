# 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib.colors import from_levels_and_colors
from cartopy import crs
from cartopy.feature import NaturalEarthFeature, COLORS
from netCDF4 import Dataset
from wrf import (getvar, to_np, get_cartopy, latlon_coords, vertcross,
                 cartopy_xlim, cartopy_ylim, interpline, CoordPair)
```

```python
# specify where is the location of the data
path_in = "data/"
path_out = "./"
# Open the NetCDF file
ncfile = Dataset(path_in + 'wrfout_d01_2016-05-09_00^%00^%00')
print(ncfile.variables.keys())
```

```python
# Get the sea level pressure
slp = getvar(ncfile, "slp")
print(slp.dims)
print(slp)
```

```python
lons_ = getvar(ncfile, "lon")
lats_ = getvar(ncfile, "lat")
print("max lon:", np.max(lons_.values))
print("min lon:", np.min(lons_.values))
print("max lat:", np.max(lats_.values))
print("min lat:", np.min(lats_.values))
print("time:", lats_.Time)
```

```python
lats, lons = latlon_coords(slp)

print(np.min(lats.values))
print(np.max(lats.values))
```
