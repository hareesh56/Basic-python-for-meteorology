## Problem:
- plot a time series of AOD
- each time step is assigned to a specific cluster (e.g. from cluster 1 to 9)
- plot a histogram under the time series, but each cluster has different color

# 1. Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors
```

# 2. Open CSV file

```python
# specify where is the location of the data
path_in = "data/"
path_out = "./"
# Open the CSV file
fname   = 'test_data.csv'
df = pd.read_csv(path_in + fname)
print(df)
print(df.keys())
```

# 3. Process and plot the data

```python
colo = ['r','b','g','m','k','c','gold','darkblue','olive']
fig = plt.figure(figsize=(12,12))
x = df['ind'].values
y = df['aod'].values
plt.bar(x, y, color ='b')
plt.plot(x, y, color ='k')

x_index = []
for i in range(1, 10):
    df_clus = df[df['cluster']==i]
    x_index_ = df_clus.index
    x_index.append(x_index_)

print('at this x_index[0] the color will be r')
print(x_index[0])

for i in range(0,9):
    plt.bar(x[x_index[i]],y[x_index[i]], color=colo[i])

# Set the x-axis and  y-axis labels
plt.xlabel("Time", fontsize=25)
plt.ylabel("AOD", fontsize=25)

# Set the range of y-axis
plt.ylim([0,150])

# Add a title
plt.title("ADO time series", {"fontsize" : 25})
plt.show()
```
