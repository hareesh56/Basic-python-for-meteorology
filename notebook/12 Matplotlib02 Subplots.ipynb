```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['font.family'] = 'serif'
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 128)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2*x) + np.sin(2*x)

ncols, nrows = 3,1

fig, ax = plt.subplots(nrows, ncols, figsize=(10, 4))

ax[0].plot(x, y1, color='r')
ax[0].axvline(x=0, color = 'k')
ax[0].axhline(y=0, color = 'k')    

ax[1].plot(x, y2, color='b')
ax[1].axvline(x=0, color = 'k')
ax[1].axhline(y=0, color = 'k')    

ax[2].plot(x, y3, color='m')
ax[2].axvline(x=0, color = 'k')
ax[2].axhline(y=0, color = 'k')    
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 128)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2*x) + np.sin(2*x)

ncols, nrows = 3,1

fig, ax = plt.subplots(nrows, ncols, figsize=(10, 4))

color  = ('r', 'b', 'm')
y_     = (y1, y2, y3)

for i in range(ncols):
    ax[i].plot(x, y_[i], color=color[i])
    ax[i].axvline(x=0, color = 'k')
    ax[i].axhline(y=0, color = 'k')    
```

```python
ncols, nrows = 3, 3

fig, axes = plt.subplots(nrows, ncols, figsize=(10, 8))

for i in range(nrows):
    for j in range(ncols):
        axes[i, j].set_xticks([])
        axes[i, j].set_yticks([])
        axes[i, j].text(0.6, 0.5, "axes[%d, %d]" % (i, j),
                        horizontalalignment='right')
```

```python
fig = plt.figure()

def remove_ticklabels(ax):
    ax.set_yticklabels([])
    ax.set_xticklabels([])

ax0 = plt.subplot2grid((2, 3), (0, 0))
ax1 = plt.subplot2grid((2, 3), (0, 1))
ax2 = plt.subplot2grid((2, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((2, 3), (0, 2), rowspan=2)

axes = [ax0, ax1, ax2, ax3]
[ax.text(0.5, 0.5, "ax%d" % i, horizontalalignment='center') for i, ax in enumerate(axes)]
[remove_ticklabels(ax) for ax in axes]
```

```python
fig = plt.figure()

def remove_ticklabels(ax):
    ax.set_yticklabels([])
    ax.set_xticklabels([])

ax0 = plt.subplot2grid((3, 3), (0, 0))
ax1 = plt.subplot2grid((3, 3), (0, 2))
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (2, 1))
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

axes = [ax0, ax1, ax2, ax3, ax4]
[ax.text(0.5, 0.5, "ax%d" % i, horizontalalignment='center') for i, ax in enumerate(axes)]
[remove_ticklabels(ax) for ax in axes]
```

```python
fig = plt.figure(figsize=(4, 4))

so = mpl.gridspec.GridSpec(2, 2,
                           width_ratios=[4, 1],
                           height_ratios=[1, 4],
                           wspace=0.1, hspace=0.5
                           )

ax0 = fig.add_subplot(so[1, 0])
ax1 = fig.add_subplot(so[0, 0])
ax2 = fig.add_subplot(so[1, 1])
ax3 = fig.add_subplot(so[0, 1])

def remove_ticklabels(ax):
    ax.set_yticklabels([])
    ax.set_xticklabels([])

axes = [ax0, ax1, ax2]
[ax.text(0.5, 0.5, "ax%d" % i, horizontalalignment='center') for i, ax in enumerate(axes)]
[remove_ticklabels(ax) for ax in axes]
```
