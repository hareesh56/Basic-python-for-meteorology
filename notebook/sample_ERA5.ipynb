```python
import xarray as xr
import glob
import numpy as np
import pandas as pd
import os
```

```python
# Open the ERA5
# DATA is downloded at https://cds.climate.copernicus.eu/cdsapp#!/dataset/reanalysis-era5-pressure-levels?tab=form
path = os.getcwd()
all_fname = glob.glob(path + "/*.nc")
fname = xr.open_mfdataset(all_fname)
fname
```

```python
def find_nearest_grid_point(era_data, time, lat, lon):
    '''
    Input:
    era_data: the era5 netcdf file
    time, lat, lon is in the format as follows
        time: '2018-02-15T06:06:00'
        lat: '-55.5555123123'
        lon: '155.55512323'
    Output: 
    ERA5 profile at a grid point with 37 levels
    '''
    return era_data.t.sel(time=time,latitude=lat,longitude=lon,method='nearest')
```

```python
target_time = '2018-02-15T12:06:00'
target_lat  = '-53.7555123123'
target_lon  = '153.21512323'

ERA5_profile = find_nearest_grid_point(era_data=fname, time=target_time, lat=target_lat, lon=target_lon)
ERA5_profile
```
